```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder,OneHotEncoder
import pickle
```

```python
#Load trained model, scaler pickle, one hot encoder
from keras.models import load_model


model = load_model("model.h5")

with open("scaler.pkl","rb") as file:
  scaler = pickle.load(file)
with open("One_hot_encoder.pkl", "rb") as file:
  one_hot_encoder = pickle.load(file)
with open("label_encoder.pkl", "rb") as file:
  label_encoder = pickle.load(file)
```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
# encode the actual Geography value from input_data without refitting the encoder

geo_encoded = one_hot_encoder.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=one_hot_encoder.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_data = pd.DataFrame([input_data])
input_data = pd.concat([input_data.drop("Geography",axis = 1),geo_encoded_df],axis =1)
```

```python
input_data
```

```python
input_data["Gender"] = label_encoder.transform(input_data["Gender"])
```

```python
input_data
```

```python
input_data = scaler.transform(input_data)
```

```python
prediction = model.predict(input_data)
```

```python
prediction
```

```python
if(prediction[0][0]>0.5):
  print("Person is likely to stay in the bank")
else:
  print("Person is likely to leave the bank")
```
